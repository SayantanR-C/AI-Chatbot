```python

!pip install nltk --quiet
import json
import numpy as np
import nltk
nltk.download('punkt')
from nltk.stem import WordNetLemmatizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder


with open('intents.json') as file:
    data = json.load(file)

lemmatizer = WordNetLemmatizer()


sentences = []
labels = []

for intent in data['intents']:
    for pattern in intent['patterns']:
        sentences.append(pattern)
        labels.append(intent['tag'])


tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(sentences)
padded_sequences = pad_sequences(sequences, padding='post')

lbl_encoder = LabelEncoder()
lbl_encoder.fit(labels)
encoded_labels = lbl_encoder.transform(labels)

model = Sequential()
model.add(Embedding(input_dim=len(word_index)+1, output_dim=128, input_length=padded_sequences.shape[1]))
model.add(LSTM(128))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dense(len(set(encoded_labels)), activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

model.fit(padded_sequences, np.array(encoded_labels), epochs=200, batch_size=8)

model.save("chatbot_lstm_model.h5")
import pickle
pickle.dump(tokenizer, open('tokenizer.pkl', 'wb'))
pickle.dump(lbl_encoder, open('label_encoder.pkl', 'wb'))
```

```python
import random
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences

model = load_model("chatbot_lstm_model.h5")
tokenizer = pickle.load(open('tokenizer.pkl', 'rb'))
lbl_encoder = pickle.load(open('label_encoder.pkl', 'rb'))

def chat():
    print("🤖 ChatBot Pro is ready! Type 'quit' to exit.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "quit":
            break

        sequence = tokenizer.texts_to_sequences([user_input])
        padded = pad_sequences(sequence, maxlen=20, padding='post')
        prediction = model.predict(padded)[0]
        tag = lbl_encoder.inverse_transform([np.argmax(prediction)])[0]

        for intent in data['intents']:
            if intent['tag'] == tag:
                print("Bot:", random.choice(intent['responses']))
                break

# Start chatting
chat()
```
